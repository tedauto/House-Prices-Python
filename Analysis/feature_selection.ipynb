```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#for feature selection will use sklearn

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromMode
```

```python

data = pd.read_csv('X_train.csv')

data.head()
```

```python
#Capture the dependent feature
y_train=data[['SalePrice']]
```

```python
#drop the dependent feature i.e id and saleprice
X_train=data.drop(['Id','SalePrice'],axis=1)
```

```python
#Apply Feature Selection
# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) #  set the seed, the random state in this function
feature_sel_model.fit(X_train, y_train)
```
