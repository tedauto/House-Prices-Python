```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

## Missing values

```python
## handling missing values in categorical features
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype=='O']
for feature in features_nan:
    print('{} :{}% missing values'.format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
#replacing null values with a  new label
def replace_cat_feature(df,features_nan):
    data=df.copy()
    data[features_nan]=data[features_nan].fillna("Missing")
    return data

df=replace_cat_feature(df,features_nan)

df[features_nan].isnull().sum()
    
```

```python
df.head()
```

```python
##handling missing values in numerical variables
numerical_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

for feature in  numerical_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
##replacing the missing values
#since there are outliers as seen in the exploratory data analysis will use median
for feature in numerical_nan:
    median_value=df[feature].median()
    
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace =True)
    
#check if there is any null values
df[numerical_nan].isnull().sum()
```

```python
df.head(50)
```

```python
##handing date time variables
#converting then to years
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]
```

```python
#this shows the number of years subtracted from the YrSold
df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
#from the EDA i found out that the numerical variables were skewed 
#should convert them to log normal distribution 
num_features=['LotFrontage','LotArea','1stFlrSF','GrLivArea']

for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

```python
##handling rare categoriacl feature
#removing categorical features with less than 1% of the total observations
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
```

```python
categorical_features
```

```python
##finding % of each category with respect to the whole dataset
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'RareVar')
```

```python
df.head(50)
```

```python
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

## Feature scaling

```python
feature_scale=[feature for feature in df.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[feature_scale])

```

```python
scaler.transform(df[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([df[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(df[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv',index = True)
```
